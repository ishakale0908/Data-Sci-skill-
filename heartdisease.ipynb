```python
df=pd.read_csv("heart.csv")
df
```

```python
print("Null values: ", df.isnull().sum())
print("Duplicate values: ", df.duplicated().sum())
df.info()
df.describe()
```

```python
# Encode labels in column 'famhist'.
label_encoder=preprocessing.LabelEncoder()
df['famhist']=label_encoder.fit_transform(df['famhist'])
df.info()
df
```

```python
sns.set(style='whitegrid', context='notebook')
cols = ['sbp','tobacco','ldl','adiposity','famhist','typea','obesity', 'alcohol','age', 'chd']
```

```python
import numpy as np

f, ax = plt.subplots(figsize=(15, 10))
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.5)
hm = sns.heatmap(cm,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size': 15},
                 yticklabels=cols,
                 xticklabels=cols)

plt.show()
```

```python
X=df[['tobacco','ldl','adiposity','famhist','typea','obesity', 'alcohol','age']].values
y=df[['chd']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model=LogisticRegression(C=1, penalty='l2')
model.fit(X_train,y_train)
```

```python
print ('Training Accuracy: %.2f' % model.score(X_train,y_train))
print ('Test Accuracy: %.2f' % model.score(X_test,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model1=GaussianNB()
model1.fit(X_train,y_train)
```

```python
print ('Training Accuracy: %.2f' % model1.score(X_train,y_train))
print ('Test Accuracy: %.2f' % model1.score(X_test,y_test))
```

```python
expected = y_test
predicted = model1.predict(X_test)
```

```python

print(classification_report(expected, predicted))
print(confusion_matrix(expected, predicted))
```
